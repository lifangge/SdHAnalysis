# SdH Analysis
----------------------------
### Example of the analysis of Shubnikov-de Haas data measured using TDO in pulsed fields.


```python
import sdhanalysis as sdh
%matplotlib inline
```

## Load data:

```python
sdh15_3 = sdh.sdh_load(15, 3, plot=True)
```

```python
sdh15_4 = sdh.sdh_load(15, 4, plot=True)
```

```python
sdh15_5 = sdh.sdh_load(15,5, plot=True)
```

-------------------

## Subtract non-oscillatory background:

```python
deg, Bmin, Bmax = 7, 1, 2.5
sdh15_3 = sdh.back_subtract(sdh15_3, deg, Bmin, Bmax, yscale=1.2e4, save=True)
```

```python
deg, Bmin, Bmax = 5, 1, 4
sdh15_4 = sdh.back_subtract(sdh15_4, deg, Bmin, Bmax, yscale=7e4, save=True)
```

```python
deg, Bmin, Bmax = 7, 1, 5
sdh15_5 = sdh.back_subtract(sdh15_5, deg, Bmin, Bmax, yscale=8e4, save=True)
```

-----------------

## Identify component frequencies:

```python
sdh15_3 = sdh.get_fft_peaks(sdh15_3, direc='Down', mph=7e5, mpd=10, threshold=1, xmax=600, keep_ind=None, save=True)
```

```python
sdh15_4 = sdh.get_fft_peaks(sdh15_4, mph=1.2e6, mpd=10, threshold=1, xmax=400, keep_ind=[0,2,3,4,6], save=True)
```

```python
sdh15_5 = sdh.get_fft_peaks(sdh15_5, mph=1.3e6, mpd=10, threshold=1, xmax=400, keep_ind=[1,3,4,6,8], save=True)
```

-----------------------

## Filter to isolate specific orbits:

```python
center_freq, passband, orbit = 60, 10, '60 T'
sdh15_5 = sdh.isolate_orbit(sdh15_5, center_freq, passband, orbit, order=2, save=True)
```

```python
center_freq, passband, orbit = 118, 10, '118 T'
sdh15_5 = sdh.isolate_orbit(sdh15_5, center_freq, passband, orbit, order=2, save=True)
```

```python
center_freq, passband, orbit = 324, 10, '324 T'
sdh15_5 = sdh.isolate_orbit(sdh15_5, center_freq, passband, orbit, order=2, save=True)
```

```python
center_freq, passband, orbit = 384, 10, '384 T'
sdh15_5 = sdh.isolate_orbit(sdh15_5, center_freq, passband, orbit, order=2, save=True)
```

```python
center_freq, passband, orbit = 446, 10, '446 T'
sdh15_5 = sdh.isolate_orbit(sdh15_5, center_freq, passband, orbit, order=2, save=True)
```
